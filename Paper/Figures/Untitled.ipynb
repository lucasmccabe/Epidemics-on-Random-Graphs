```python
import plotnine as p9
from plotnine import aes, scale_colour_manual
import pandas as pd
import math
```

```python
def z(t, pλ, n, d, α, ϕ):
    '''
    t = time step
    pλ = p_adjacent times p_infect
    n = population
    d = delay to begin rollout
    α = rollout rate
    ϕ = vaccine effectiveness
    '''
    if t==1:
        return min(n, 1)
    elif 1<t<=d:
        return min(n, 1+(t-1)*n*pλ)
    elif d<t<=(d+1.0/α):
        return min(n, 1+n*pλ*((d-1)+(t-d)*(1-α*(t-d)*ϕ)))
    else:
        return min(n, 1+n*pλ*((d-1)+(1.0/α)*(1-ϕ)+(t-1.0/α)*(1-ϕ)))
```

```python
t = 20 #time step
pλ = 0.1 #p_adjacent times p_infect
n = 1000 #population
d = 5 #delay to begin rollout
α = 0.5 #rollout rate
ϕ = 0.8 #vaccine effectiveness
z(t, pλ, n, d, α, ϕ)
```

```python
t = 20 #time step
pλ = 0.1 #p_adjacent times p_infect
n = 1000 #population
d = 5 #delay to begin rollout
α = 0.1 #rollout rate

ax = (p9.ggplot(pd.DataFrame(data={"x": [1, 25]}), aes(x="x"))
    + p9.stat_function(fun=lambda x: z(x, pλ, n, 50, α, 1))
    + p9.stat_function(fun=lambda x: z(x, pλ, n, d, α, 1))
    + p9.stat_function(fun=lambda x: z(x, pλ, n, d, α, 0.75))
    + p9.stat_function(fun=lambda x: z(x, pλ, n, d, α, 0.5))
    + p9.stat_function(fun=lambda x: z(x, pλ, n, d, α, 0.25))
    + p9.xlab("Time Step (t)")
    + p9.ylab("Upper Bound on E[Z_t]")
    + p9.ggtitle("A Bad Figure That I Need to Re-Do")
)

print(ax)
```
