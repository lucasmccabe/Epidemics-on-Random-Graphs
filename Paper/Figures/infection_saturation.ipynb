```python
import plotnine as p9
from plotnine import aes
import pandas as pd
import math
```

```python
n=1000

ax = (p9.ggplot(pd.DataFrame(data={"x": [0.001, 0.998]}), aes(x="x"))
    + p9.stat_function(fun=lambda x: math.log(((n-1)/x)-n)+1)
    + p9.xlab("pλ")
    + p9.ylab("t (Time Steps)")
    + p9.ggtitle("Lower Bound for S(1000, p, λ, μ)")
)

print(ax)
```
